```python
(
  df
 .groupby("param_marka-pojazdu")["price_value"]
 .agg(np.mean)
 .sort_values(ascending=False)
 .head(50)
).plot(kind="bar", figsize=(20, 5))
```

```python
(
  df
 .groupby("param_marka-pojazdu")["price_value"]
 .agg(np.median)
 .sort_values(ascending=False)
 .head(50)
).plot(kind="bar", figsize=(20, 5))
```

```python
(
  df
 .groupby("param_marka-pojazdu")["price_value"]
 .agg([np.mean, np.median, np.size])
 .sort_values(by="mean", ascending=False)
 .head(50)
).plot(kind="bar", figsize=(20, 5), subplots=True)
```

```python
def group_and_barplot(feat_groupby, feat_agg="price_value", agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True, fig_size=(15, 5)):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(feat_sort, ascending=False)
    .head(top)
  ).plot(kind="bar", figsize=fig_size, subplots=subplots)
```

```python
group_and_barplot("param_marka-pojazdu")
```

```python
group_and_barplot("param_kraj-pochodzenia", feat_sort="size")
```

```python
group_and_barplot("param_kolor", fig_size=(10, 5))
```

```python
group_and_barplot("param_liczba-drzwi", fig_size=(5, 5))
```
